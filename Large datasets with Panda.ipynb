```python
import pandas as pd

def preprocess_patetnt(in_f, out_f, size, use_cols):
    reader = pd.read_table(in_f, sep=',', usecols=use_cols, chunksize=size)
    for chunk in reader:
        chunk.columns = use_cols
        result = chunk
        result.to_csv(out_f, index=False, header=False, mode='w')

i=0
y=1
filepath='Desktop/didelphis_first10000.txt'
size=5000

while i < 10001:
    V1 = str(i)
    V2 = str(i+1)
    V3 = str(i+2)
    use_cols= [V1, V2, V3]
    outpath='Desktop/didelphis/didelphis'+str(y)+'.txt'
    print(outpath)
    preprocess_patetnt(filepath,outpath,size, use_cols)
    i=i+3
    y=y+1
```
